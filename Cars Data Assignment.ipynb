# 1. Read the dataset to the python environment.

```python
#Import the main libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Read the csv to a variable
carslist=pd.read_csv("cars_data.csv",sep='\t')
#Display the variable and see if the data is visible as rows and colums
carslist
```

# 2.Check for the null values present in the dataset.


```python
#Use the command isnull to check for the presence of null columns
nullcols=carslist.columns[carslist.isnull().any()]
#Display the nullcols value and see if there are any nulls
nullcols
```

```python
#Display the presence of null columns in spreadsheet using a true/false statement
print(" presence of null values in the cars spreadsheet:",carslist.isnull().values.any())
```

# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
# Use a simple bar graph plot for the column "Car Gender" in x axis and frequency in y axis 
plt.figure(figsize=(6,6))
plt.hist(carslist["Car Gender"],color='magenta',rwidth=1.5)
plt.title('Male vs female buyers in sales',fontsize=12)
plt.xlabel('Car Gender')
plt.ylabel('frequency')
```

```python
#Inference
# The number of female buyers are slightly more than male buyers
```

# 4.Find the top 5 cars based on their sales price.

```python
#Method 1 - use descending sort function by setting ascending=false and get the top 5 values 
carslist.sort_values('Sale Price',ascending=False).head(5)
```

```python
#Method 2 - use nlargest function to get top 5 values 
Topfivepricedcars = carslist.nlargest(5,'Sale Price')
Topfivepricedcars
```

```python
#Inference
#By cross checking the above 2 methods it is clear that Volkswagen Jetta III, Audi A6, Lexus LX, Mazda MX-6 and Lexus ES 
#are the top 5 cars having the biggest Sale Price 
```

# 5. Find the least 5 cars based on their Resell price.


```python
#Method 1 - use ascending sort function by setting ascending=true (default) and get the least 5 values 
carslist.sort_values('Resell Price').head(5)
```

```python
#Method 2 - use nsmallest function to get the 5 least resell prices 
CheapfiveCars = carslist.nsmallest(5,'Resell Price')
CheapfiveCars
```

```python
#Inference
#By cross checking the above 2 methods it is clear that Chevrolet 1500,GMC Savana 1500, Jaguar XF, Mitsubishi 3000GT & Infiniti G37
# are the 5 cars having the least ReSell Price 
```
